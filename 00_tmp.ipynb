```python
sns.set(style='darkgrid')
df=np.random.multivariate_normal(mean=[0,1],cov=[(1,-0.4),(-0.4,1)],size=200)
df=pd.DataFrame(df,columns=['a','b'])
df.head()
```

```python
df.describe()
```

```python
sns.set(style='white')
g=sns.jointplot(data=df,x='a',y='b',kind='kde',nlevels=15,cmap='viridis')
g.plot_joint(plt.scatter,c='w',s=30,marker='+')
```
