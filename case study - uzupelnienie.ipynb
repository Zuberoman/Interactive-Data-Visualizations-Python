```python
import plotly.graph_objects as go

def fetch_financial_data(company='AMZN'):
    import pandas_datareader.data as web
    return web.DataReader(name=company,data_source='stooq')
```

```python
df_amzn=fetch_financial_data(company='AMZN')
df_aapl=fetch_financial_data(company='AAPL')
df_msft=fetch_financial_data(company='MSFT')
df_nflx=fetch_financial_data(company='NFLX')
df_twtr=fetch_financial_data(company='TWTR')
```

```python
df_amzn.head(1)
```

```python
import pandas as pd

df=pd.concat([df_amzn.Close,df_aapl.Close,df_msft.Close,df_nflx.Close,df_twtr.Close], axis=1, join='inner')
df.columns=['amazon','apple','microsoft','netflix','twitter']
df=df.reset_index()
df.head()
```

```python
df.describe()
```

```python
trace0=go.Scatter(x=df.Date,
                 y=df.amazon,
                 name='Amazon')
trace1=go.Scatter(x=df.Date,
                 y=df.apple,
                 name='Apple')
trace2=go.Scatter(x=df.Date,
                 y=df.microsoft,
                 name='Microsoft')
trace3=go.Scatter(x=df.Date,
                 y=df.netflix,
                 name='Netflix')
trace4=go.Scatter(x=df.Date,
                 y=df.twitter,
                 name='Twitter')

data=[trace0, trace1, trace2, trace3, trace4]

layout=go.Layout(yaxis_type='log')

fig=go.Figure(data=data,layout=layout)
fig.show()
```

```python
fig=go.Figure()

fig.add_trace(go.Scatter(
    x=df.Date,
    y=df.amazon,
    name='Amazon'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y=df.apple,
    name='Apple'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y=df.netflix,
    name='Netflix'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y=df.twitter,
    name='Twitter'
))
fig.add_trace(go.Scatter(
    x=df.Date,
    y=df.microsoft,
    name='Microsoft'
))
fig.update_layout(yaxis_type='log')
fig.show()
```

```python
data=[go.Scatter(x=df.Date,y=df[col],name=col) for col in df.columns[1:] ]

layout=go.Layout(yaxis_type='log')

fig=go.Figure(data=data,layout=layout)

fig.write_html('price.html')

fig.show()
```
